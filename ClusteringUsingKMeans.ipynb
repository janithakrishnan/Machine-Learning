```python
#Clustering using Kmeans on make blobs
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=200,centers=5, cluster_std=0.8,random_state=0)
print(X)
print(y)
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
plt.show()
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_kmeans,cmap='viridis')
plt.show()
```

```python
#Clustering using Kmeans on iris based on sepal leangth and sepal width
from sklearn.datasets import load_iris
iris=load_iris()
iris=iris.data
X=iris[:,:2]
import matplotlib.pyplot as plt
plt.title('Scatter plot of Sepal Length vs Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(X[:,0],X[:,1])
plt.show()
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
plt.title('Clustering based on Sepal length and width ')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(X[:,0],X[:,1],c=y_kmeans,cmap='viridis')
plt.show()
```
